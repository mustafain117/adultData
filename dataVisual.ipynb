```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", delimiter=",", header=None,
                 names=["age", "workplace", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex","capital-gain","capital-loss","hours-per-week", "native-country", "y"])

x = df['y']
ax = sns.countplot(x=x, data=df)
```

```python
def plot_data(hue, data): 
    for i,col in enumerate(df.columns):
        plt.figure(i)
        sns.set(rc={'figure.figsize':(11.7,8.27)})
        ax = sns.countplot(x=data[col],hue=hue, data=data)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

```python
plt_Data = df.drop('y', 1)
```

```python
hue = df['y']
```

```python
plot_data(hue, plt_Data)
```
